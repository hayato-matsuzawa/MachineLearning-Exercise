```python
import numpy as np
import matplotlib.pyplot as plt
```

# 課題１

$n = 1$ の 線形回帰用データ $D = (X, y)$

1. `np.random.uniform`で$X$を得る 

```python
# データ数100、特徴量1のデータを生成
X = np.random.uniform(0, 10, size=(100, 1)) # 一様分布に従う乱数を生成
X
```

2. `np.random.randn`で$w$を決める

```python
w = np.random.randn(1, 1) # 正規分布に従う乱数を生成
w
```

3. `np.random.randn`でノイズを決める

```python
e = np.random.randn(100, 1) * 0.1 # ノイズの標準偏差を0.1に設定
e 
```

4.  $y = Xw + e$ で $y$ を決める

```python
y = X @ w + e 
y 
```

評価用のデータ $E = (X, y)$

```python
X_test = np.random.uniform(0, 10, size=(100, 1))
e_test = np.random.randn(100, 1) * 0.1

y_test = X_test @ w + e_test
```

# 課題２

$D = (X, y)$ から $w$ を推定

#### 1. データ$E$と直線 $\hat{y} = x^\top\hat{w}$ を $xy$ 平面としてプロットせよ

```python
# 推定値 w_hat を得る
w_hat = np.linalg.lstsq(X, y, rcond=None)[0] #戻り値の最初の要素だけ取り出す
w_hat
```

```python
# 予測値の計算
y_pred = X_test @ w_hat
y_pred
```

```python
# 評価用データEの散布図
plt.scatter(X_test, y_test, label="Test Data (E)")

# 推定した直線のプロット
plt.plot(X_test, y_pred, color='red', label="Prediction")

plt.legend()
plt.show()
```

#### 2. $D$ のデータ数 $m$ を増やすと，$E$ での二乗誤差 $||y - \hat{y}||^2$ はどう変化するか

```python
error = np.sum((y_test - y_pred)** 2)
error
```

```python
# D のデータ数 m を変えて二乗誤差を求める
for m in [10, 100, 1000, 10000, 100000, 1000000]:
    X_ex = np.random.uniform(0, 10, size=(m, 1))
    e_ex = np.random.randn(m, 1) * 0.1
    y_ex = X_ex @ w + e_ex
    w_hat_ex = np.linalg.lstsq(X_ex, y_ex, rcond=None)[0]
    y_pred_ex = X_test @ w_hat_ex
    error_ex = np.sum((y_test - y_pred_ex) ** 2)
    print(f"m={m}, error={error_ex}")
    
```

$D$のデータ数を増やすと，二乗誤差は小さくなっていき，一定の値に収束した．

#### 3. $D$ のデータ数 $m$ を増やすと，傾き $\hat{w}$ はどのような値に近づくだろうか

```python
for m in [10, 100, 1000, 10000, 100000]:
    X_ex = np.random.uniform(0, 10, size=(m, 1))
    e_ex = np.random.randn(m, 1) * 0.1
    y_ex = X_ex @ w + e_ex
    w_hat_ex = np.linalg.lstsq(X_ex, y_ex, rcond=None)[0]
    print(f"m={m}, w_hat={w_hat_ex}")
print(f"True w: {w}")
```

$D$のデータ数を増やすと，傾き$\hat{w}$は真の$w$の値に近づいた．

# 課題３

#### `sklearn`の回帰用データ`sklearn.datasets.load_diabetes`で課題2-2, 2-3を実行

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
X_diabetes= diabetes.data
y_diabetes = diabetes.target
```

```python
# データの形状を確認
print(X_diabetes.shape)
print(y_diabetes.shape)
```

```python
# 訓練データを中心化する
X_diabetes_mean = X_diabetes.mean(axis=0)
X_diabetes_centered = X_diabetes - X_diabetes_mean

y_diabetes_mean = y_diabetes.mean()
y_diabetes_centered = y_diabetes - y_diabetes_mean
```

```python
# 先頭の300件を訓練データ、残りを評価用データとする
X_diabetes_train = X_diabetes_centered[:300]
y_diabetes_train = y_diabetes_centered[:300]

X_diabetes_test = X_diabetes_centered[300:]
y_diabetes_test = y_diabetes_centered[300:]
```

#### 訓練データ数$m$を増やして，評価用データでの二乗誤差の変化を確認

```python
for m in [10, 50, 100, 200, 300]:
    X_train_subset = X_diabetes_train[:m]
    y_train_subset = y_diabetes_train[:m]
    
    w_hat_subset = np.linalg.lstsq(X_train_subset, y_train_subset, rcond=None)[0]
    
    y_pred_subset = X_diabetes_test @ w_hat_subset
    error_subset = np.sum((y_diabetes_test - y_pred_subset) ** 2)
    
    print(f"m={m}, error={error_subset}")
```

データ数$m$を増やすと，二乗誤差は小さくなっていった．

#### 訓練データ数$m$を増やして，傾き$\hat{w}$の変化を確認

```python
for m in [10, 50, 100, 200, 300]:
    X_train_subset = X_diabetes_train[:m]
    y_train_subset = y_diabetes_train[:m]
    
    w_hat_subset = np.linalg.lstsq(X_train_subset, y_train_subset, rcond=None)[0]
    
    print(f"m={m}, w_hat={w_hat_subset}")
```

データ数$m$を増やすと，過学習が抑えられ，極端にばらついていた$\hat{w}$が安定した値に収束した．

# 考察
